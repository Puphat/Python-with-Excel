```python
price_usd = taxi.loc[:, ['fare','tip','tolls','total']]
price_th = price_usd*33.18
price_th
```

```python
price_per_distance = taxi['total']/taxi['distance']
price_per_distance
```

```python
taxi['Price_per_distance'] = price_per_distance
```

```python
taxi
```

```python
price_th.columns = ['fare_th', 'tip_th', 'tolls_th', 'total_th']
```

```python
report = taxi.merge(price_th,left_index=True,right_index=True)
report
```

```python
report.to_excel('report.xlsx')
```

```python
tips = sns.load_dataset('tips')
tips['pct_tips'] = tips['tip']/tips['total_bill']
tips
```

```python
sns.catplot(x='day',y='pct_tips',data=tips)
```

```python
sns.catplot(x='day',y='pct_tips',data=tips,kind='box')
```

```python
tips.sort_values(by='pct_tips',ascending=False)
```
